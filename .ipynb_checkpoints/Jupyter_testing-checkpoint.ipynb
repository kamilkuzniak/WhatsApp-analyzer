```python
#error handling when the date is not present
with open('WhatsApp Chat with Sara.txt', 'r') as chat:
    messages = chat.readlines()
    
timestamps = []
content = []
names = []
dates = []
hours = []
full_dates = []
hour = []

for i in range(len(messages)): 
    try:
        timestamps.append(time.strptime(messages[i].split('-')[0].split(',')[0] + messages[i].split('-')[0].split(',')[1], '%m/%d/%y %H:%M '))
    except:
        num = 1
        while True:
            if messages[i-num] != 'NaN':
                messages[i-num] += (' ' + messages[i])
                messages[i] = 'NaN'
                break
            else:
                num +=1
              
while True:
    try:
        messages.remove('NaN')
    except ValueError:
        break
```

```python
#for i in range(len(messages)):
#for i in range(1):
for message in messages:
    try:
        content.append(message.split('-')[1].split(':',maxsplit=1)[1].replace(' ', '',1))
        names.append(message.split('-')[1].split(':',maxsplit=1)[0].replace(' ', '',1))
    except:
        content.append(message.split('-')[1].replace(' ', '',1))
        names.append('-')
```

```python
for timestamp in timestamps:
    full_dates.append(datetime.fromtimestamp(mktime(timestamp)))
    dates.append(datetime.fromtimestamp(mktime(timestamp)).strftime('%d/%m/%Y'))
    hours.append(datetime.fromtimestamp(mktime(timestamp)).strftime('%H:%M'))
```

```python
for timestamp in timestamps:
    full_dates.append(datetime.fromtimestamp(mktime(timestamp)))
    dates.append(datetime.fromtimestamp(mktime(timestamp)).date())
    hours.append(datetime.fromtimestamp(mktime(timestamp)).time())
    hour.append(datetime.fromtimestamp(mktime(timestamp)).hour)
```
